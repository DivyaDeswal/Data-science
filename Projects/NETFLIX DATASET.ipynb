33. Bin the Duration column into categories: "Short", 
"Medium", "Long"

```python
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(float).astype('Int64')
df.head()
```

```python
def duration(d): 
    if d < 50:
        return 'Short'
    elif 50 <= d < 110:
        return 'Medium'
    else:
        return 'Long'
df['Duration_Category'] = df['Duration'].apply(duration)
print(df[['Title', 'Duration', 'Duration_Category']].head())
```
