## Custom Functions:
38. Write a function to categorize "Age" into "Young", "Adult", and "Senior" and apply it to the dataset.

```python
def categorize_age(age):
    if age < 18:
        return "Young"
    elif 18 <= age < 60:
        return "Adult"
    else:
        return "Senior"
df['Age Category'] = df['Age'].apply(categorize_age)
print(df['Age Category'])
```

39. Create a function that labels purchases as "Expensive" if the "Purchase Amount (USD)" is above $100.

```python
def label_expensive(purchase_amount):
    if purchase_amount > 100:
        return "Expensive"
    else:
        return "Affordable"
df['Purchase Label'] = df['Purchase Amount (USD)'].apply(label_expensive)
print(df['Purchase Label'])
```

40. Define a function to calculate loyalty status based on "Previous Purchases" and create a new column with this status.


```python
def calculate_loyalty(previous_purchases):
    if previous_purchases > 50:
        return "Platinum"
    elif 20 <= previous_purchases <= 50:
        return "Gold"
    else:
        return "Silver"
df['Loyalty Status'] = df['Previous Purchases'].apply(calculate_loyalty)
print(df['Loyalty Status'])
```
