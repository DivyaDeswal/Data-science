30.Add a new column that categorizes salaries into 'Low', 'Medium', and 'High' 
based on defined salary ranges.

```python
def categorize_salary(salary):
    if salary < 30000:
        return 'Low'
    elif 50000 <= salary <= 100000:
        return 'Medium'
    else:
        return 'High'
        
df['salary_category'] = df['salary_in_usd'].apply(categorize_salary)
df[['salary_in_usd', 'salary_category']].head()
```
